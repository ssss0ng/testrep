```python
# -------------------------------
# Python List Methods Examples
# -------------------------------

# Create an initial list
my_list = ["apple", "banana", "cherry"]
print("Original list:", my_list)


# ---------------- append() ----------------
# Adds an element at the end of the list
my_list.append("orange")              # Add "orange" to the end
print("After append:", my_list)


# ---------------- clear() ----------------
# Removes all elements from the list
temp_list = my_list.copy()            # Copy list to avoid losing original data
temp_list.clear()                     # Remove all elements
print("After clear:", temp_list)


# ---------------- copy() ----------------
# Returns a copy of the list
copied_list = my_list.copy()          # Create a duplicate list
print("Copied list:", copied_list)


# ---------------- count() ----------------
# Returns number of elements with specified value
count_apple = my_list.count("apple")  # Count occurrences of "apple"
print("Count of apple:", count_apple)


# ---------------- extend() ----------------
# Add elements of another iterable to the list
extra_items = ["grape", "mango"]      # Another list
my_list.extend(extra_items)           # Add elements to original list
print("After extend:", my_list)


# ---------------- index() ----------------
# Returns index of first matching element
index_banana = my_list.index("banana")  # Find position of "banana"
print("Index of banana:", index_banana)

# ---------------- insert() ----------------
# Adds element at specific position
my_list.insert(1, "kiwi")             # Insert "kiwi" at index 1
print("After insert:", my_list)


# ---------------- pop() ----------------
# Removes element at specified position
removed_item = my_list.pop(2)         # Remove item at index 2
print("Removed item using pop:", removed_item)
print("After pop:", my_list)


# ---------------- remove() ----------------
# Removes item with specified value
my_list.remove("apple")               # Remove first occurrence of "apple"
print("After remove:", my_list)


# ---------------- reverse() ----------------
# Reverses the order of the list
my_list.reverse()                     # Reverse list order
print("After reverse:", my_list)

# ---------------- sort() ----------------
# Sorts the list (ascending by default)
numbers = [5, 2, 9, 1]                # Create a numeric list
numbers.sort()                        # Sort numbers in ascending order
print("Sorted numbers:", numbers)

# ---------------- sort() (reverse order) ----------------
# Sort list in descending order
numbers.sort(reverse=True)            # Sort in descending order
print("Sorted numbers (descending):", numbers)
```

```python
colors = ["red", "green", "blue", "yellow"]

print(colors[-1])     # Negative index: access last item → yellow
print(colors[-2])     # Second last item → blue

print(colors[1:3])    # Slice from index 1 to 2 (end is excluded) → ['green', 'blue']
print(colors[:2])     # From beginning up to index 1 → ['red', 'green']
print(colors[2:])     # From index 2 to end → ['blue', 'yellow']

print(colors[::-1])   # Reverse the entire list → ['yellow', 'blue', 'green', 'red']
```

```python
# Creating a tuple (stores multiple items in one variable)
my_tuple = ("laptop", "phone", "phone", "tablet")

# Printing the whole tuple
print(my_tuple)  # Output: ('laptop', 'phone', 'phone', 'tablet')

# Tuples are ordered → we can access items using index
print(my_tuple[0])  # First item (index 0) → laptop
print(my_tuple[1])  # Second item (index 1) → phone

# Tuples allow duplicate values
print(my_tuple[1])  # phone appears more than once

# Finding total number of items in tuple
print(len(my_tuple))  # Output: 4

# Negative indexing (access from end)
print(my_tuple[-1])  # Last item → tablet

# Slicing (range of items): Slicing: tuple[start:end], (start included, end excluded)
print(my_tuple[1:3])  # Starts from index 1 (included) , Ends before index 3 (excluded), So it returns items at index 1 and 2

# Loop through tuple
for item in my_tuple:
    print(item)

# Tuple is unchangeable (immutable)
# The below line will give an error if uncommented
# my_tuple[1] = "camera"  # Not allowed

# Tuple with different data types
mixed_tuple = ("laptop", 2024, True, 99.5)

# Printing mixed data types tuple
print(mixed_tuple)  # Output: ('laptop', 2024, True, 99.5)

# Single item tuple (comma is important)
single_tuple = ("laptop",)

# Checking type to confirm it's a tuple
print(type(single_tuple))  # Output: <class 'tuple'>
```

```python
# Creating a set (stores unique items, unordered)
my_set = {"laptop", "phone", "tablet"}

# Printing the set (order may change because sets are unordered)
print(my_set)

# Loop through set items (no indexing in sets)
for item in my_set:
    print(item)  # Prints each item one by one

# Check if an item exists in the set
print("phone" in my_set)  # True if present

# Check if an item is NOT present
print("camera" not in my_set)  # True if not present

# Adding a single item to the set
my_set.add("camera")  # Adds new item

# Print updated set
print(my_set)

# Adding multiple items using another set
new_items = {"keyboard", "mouse"}

# Update set with new items
my_set.update(new_items)

# Print updated set
print(my_set)

# Adding items from a list (any iterable)
extra_items = ["monitor", "speaker"]

# Update set with list items
my_set.update(extra_items)

# Print updated set
print(my_set)

# Removing an item (will give error if item not found)
my_set.remove("phone")

# Print updated set
print(my_set)

# Removing an item safely (no error if item not found)
my_set.discard("tablet")

# Print updated set
print(my_set)

# Loop again after updates
for item in my_set:
    print(item)
```

```python
# Creating sets
set1 = {"laptop", "phone", "tablet"}
set2 = {"phone", "camera", "monitor"}

# union() → combines both sets (new set)
result_union = set1.union(set2)
print(result_union)

# | operator → same as union()
result_operator = set1 | set2
print(result_operator)

# update() → adds set2 into set1 (modifies original set)
set1.update(set2)
print(set1)

# intersection() → keeps only common items
set3 = {"laptop", "phone", "tablet"}
set4 = {"phone", "camera"}
print(set3.intersection(set4))  # common items

# difference() → items in set3 but not in set4
print(set3.difference(set4))

# symmetric_difference() → items not common in both sets
print(set3.symmetric_difference(set4))

# Joining set with tuple (only works with union)
extra_items = ("keyboard", "mouse")
print(set3.union(extra_items))  # tuple added to set
```

```python
# Creating a frozenset (immutable set)
my_frozenset = frozenset({"laptop", "phone", "tablet"})

# Printing the frozenset
print(my_frozenset)  # Output: frozenset({'laptop', 'phone', 'tablet'})

# Checking the type
print(type(my_frozenset))  # Output: <class 'frozenset'>

# Loop through frozenset (allowed)
for item in my_frozenset:
    print(item)

# Check if an item exists
print("phone" in my_frozenset)  # True if present

# Frozenset is immutable (cannot add or remove items)
# The below lines will give error if uncommented

# my_frozenset.add("camera")      # Not allowed
# my_frozenset.remove("phone")    # Not allowed
```

```python
# Creating sets
set1 = {"laptop", "phone", "tablet"}
set2 = {"phone", "camera"}

# add() → add an element
set1.add("monitor")
print(set1)

# remove() → removes item (error if not found)
set1.remove("tablet")
print(set1)

# discard() → safe remove (no error)
set1.discard("speaker")  # item not present
print(set1)

# pop() → removes random item
set1.pop()
print(set1)

# copy() → create a copy
set_copy = set1.copy()
print(set_copy)

# clear() → remove all items
temp_set = {"keyboard", "mouse"}
temp_set.clear()
print(temp_set)  # empty set

# union() → combine sets (new set)
print(set1.union(set2))

# update() → modify original set
set1.update(set2)
print(set1)

# intersection() → common items
set3 = {"laptop", "phone"}
set4 = {"phone", "camera"}
print(set3.intersection(set4))

# difference() → items in set3 not in set4
print(set3.difference(set4))

# symmetric_difference() → non-common items
print(set3.symmetric_difference(set4))

# issubset() → check subset
print(set3.issubset(set1))

# issuperset() → check superset
print(set1.issuperset(set3))

# isdisjoint() → no common items
print(set3.isdisjoint({"keyboard", "mouse"}))
```

```python
my_set = {"laptop", "phone", "tablet"}

my_set.add("camera")       # add() → inserts new item into set
print(my_set)              # camera is now in the set

my_set.remove("phone")     # remove() → deletes item, ERROR if not found
print(my_set)              # phone is removed

my_set.discard("speaker")  # discard() → safe delete, NO error if item missing
print(my_set)              # no change, speaker was not in set

# Key difference:
# remove("missing")  → raises KeyError
# discard("missing") → does nothing, no crash
```

```python
# Creating a dictionary (student information)
student = {
    "name": "Minho",          # key:value pair (string)
    "age": 22,                # integer value
    "city": "Seoul"           # another key:value pair
}

# Printing the dictionary
print(student)

# Accessing values using key
print(student["name"])       # Access using key → Minho

# Access using get() method
print(student.get("age"))    # Safer way to access → 22

# Get all keys
print(student.keys())        # Shows all keys

# Get all values
print(student.values())      # Shows all values

# Adding a new item
student["major"] = "AI"      # Adds new key:value pair

# Updating value
student["age"] = 23          # Change existing value

# Update using update() method
student.update({"city": "Busan"})

# Print updated dictionary
print(student)

# Removing an item using pop()
student.pop("major")         # Removes 'major'

# Removing last inserted item
student.popitem()

# Print after removal
print(student)

# Loop through keys
for key in student:
    print(key)               # Prints keys

# Loop through values
for value in student.values():
    print(value)             # Prints values

# Loop through both key and value
for key, value in student.items():
    print(key, value)        # Prints key + value

# Copy dictionary
student_copy = student.copy()
print(student_copy)

# Nested dictionary (multiple students)
students = {
    "student1": {
        "name": "Jisoo",
        "age": 21
    },
    "student2": {
        "name": "Hyunwoo",
        "age": 24
    }
}

# Access nested value
print(students["student2"]["name"])  # Hyunwoo

# Loop through nested dictionary
for key, value in students.items():
    print(key)                       # student1, student2
    for k in value:
        print(k, ":", value[k])      # inner dictionary items

# Clear dictionary (remove all items)
student.clear()
print(student)  # Empty dictionary
```

```python
# Creating a dictionary (student info)
student = {
    "name": "Jisoo",       # key:value pair
    "age": 21,
    "city": "Seoul"
}

# get() → access value safely
print(student.get("name"))  # Output: Jisoo

# keys() → get all keys
print(student.keys())

# values() → get all values
print(student.values())

# items() → get key-value pairs
print(student.items())

# copy() → create a copy of dictionary
student_copy = student.copy()
print(student_copy)

# setdefault() → get value, or add if not exists
student.setdefault("major", "Computer Science")  # adds if missing
print(student)

# update() → update or add new key-value pairs
student.update({"age": 22, "country": "Korea"})
print(student)

# pop() → remove item using key
student.pop("city")
print(student)

# popitem() → remove last inserted item
student.popitem()
print(student)

# fromkeys() → create new dictionary with same keys
keys = ["id", "grade", "status"]
new_dict = dict.fromkeys(keys, "N/A")
print(new_dict)

# clear() → remove all items
student.clear()
print(student)  # empty dictionary
```

```python
# Python File Handling Example for Google Colab
# This code creates, reads, appends, overwrites, and deletes a file

# Import os module for file and folder operations
import os

# Show current working directory
print("Current folder:", os.getcwd())

# ---------------- CREATE / WRITE FILE ----------------

# Open file in write mode ("w")
# If the file does not exist, Python will create it
# If the file already exists, its old content will be overwritten
with open("data.txt", "w") as f:
    f.write("Hello from Seoul\n")   # Write first line into file
    f.write("This is a file handling example.\n")  # Write second line

# Print confirmation message
print("File created and written successfully.")

# ---------------- CHECK FILE EXISTS ----------------

# Check whether the file exists in current folder
if os.path.exists("data.txt"):
    print("data.txt exists.")
else:
    print("data.txt does not exist.")

# Show all files in current folder
print("Files in current folder:", os.listdir())

# ---------------- READ FILE ----------------

# Open file in read mode ("r")
with open("data.txt", "r") as f:
    print("\nReading full file content:")
    print(f.read())   # Read and print full file content

# ---------------- READ ONE LINE ----------------

# Open file again in read mode
with open("data.txt", "r") as f:
    print("Reading first line only:")
    print(f.readline())   # Read only the first line

# ---------------- APPEND FILE ----------------

# Open file in append mode ("a")
# This adds new content at the end of the file
with open("data.txt", "a") as f:
    f.write("This line was added later.\n")   # Add new line

# Read file after appending
with open("data.txt", "r") as f:
    print("After appending:")
    print(f.read())

# ---------------- OVERWRITE FILE ----------------

# Open file in write mode again
# This will replace old content with new content
with open("data.txt", "w") as f:
    f.write("Old content was replaced.\n")   # New file content

# Read file after overwriting
with open("data.txt", "r") as f:
    print("After overwriting:")
    print(f.read())

# ---------------- CREATE NEW FILE USING x MODE ----------------

# Create a new file only if it does not already exist
# This avoids FileExistsError
if not os.path.exists("newfile.txt"):
    with open("newfile.txt", "x") as f:
        f.write("This is a newly created file.\n")
    print("newfile.txt created successfully.")
else:
    print("newfile.txt already exists.")

# ---------------- DELETE FILE ----------------

# Delete newfile.txt if it exists
if os.path.exists("newfile.txt"):
    os.remove("newfile.txt")   # Remove the file
    print("newfile.txt deleted successfully.")
else:
    print("newfile.txt does not exist.")

# ---------------- FINAL FILE LIST ----------------

# Show remaining files in current folder
print("Final files in folder:", os.listdir())
```

```python
# Import required modules
import os

# ---------------- CREATE FOLDER ----------------

folder_name = "my_data"  # folder name

# Create folder if it does not exist
if not os.path.exists(folder_name):
    os.mkdir(folder_name)   # create folder
    print("Folder created")
else:
    print("Folder already exists")

# ---------------- CREATE FILES ----------------

# File paths
txt_file = os.path.join(folder_name, "info.txt")
csv_file = os.path.join(folder_name, "data.csv")
log_file = os.path.join(folder_name, "log.txt")

# Create and write into .txt file
with open(txt_file, "w") as f:
    f.write("Name: Minho\n")       # write text
    f.write("Role: Developer\n")

# Create and write into .csv file
with open(csv_file, "w") as f:
    f.write("id,name,score\n")     # CSV header
    f.write("1,Jisoo,90\n")
    f.write("2,Hyunwoo,85\n")

# Create and write into .log file
with open(log_file, "w") as f:
    f.write("System started\n")
    f.write("No errors found\n")

print("Files created and written successfully")

# ---------------- READ ALL FILES ----------------

print("\nReading all files:\n")

# Loop through all files in folder
for file in os.listdir(folder_name):

    file_path = os.path.join(folder_name, file)  # full path

    # Open each file in read mode
    with open(file_path, "r") as f:
        print("File:", file)       # print file name
        print(f.read())            # print content
        print("-------------------")
```
