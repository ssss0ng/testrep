```python
print("="*40)
project=("OpenTrack", "1.0", 2026, "python","Jiwoo Song")
print(f"{project[0]} — Open Source Project")
print(f"Projest Lead : {project[4]}")
print("="*40)
print(f"Name : {project[0]}  |  Version : {project[1]}  |  Started : {project[2]}")
print(f"First 3 fields : {project[0:3]}")
print(f"Language count : {project.count("python")}  Language index : {project.index("python")}")
#project[1]="2.0" -> TypeError: 'tuple' object does not support item assignment
print("-"*40)
contributors=[]
for i in range(4):
  name=input("name : ")
  role=input("role : ")
  language=input("language : ")
  commits=input("commits : ")
  country=input("country : ")
  contributors.append({"name":name, "role":role, "language":language, "commits":commits, "country":country})
names=[]
for i in range(4):
  names.append(contributors[i]["name"])
names.sort()
print("-"*40)
print(f"Sorted names : {names}")
print(f"Last name    : {names[-1]}   First two : {names[:2]}")
for i in range(4):
  contributors[i].update({"status":"Active"})
print(f"Contributor 1 status : {contributors[0].get("status")}")
backup1=contributors[0].copy()
print(f"Backup : {backup1}")
print("-"*40)

issues=[]
for i in range(5):
  id=input("id : ")
  title=input("title : ")
  type=input("type(Bug/Feature) : ")
  priority=input("priority(Critical/High/Medium/Low) : ")
  reporter=input("reporter : ")
  status=input("status(Open/In Progress/Resolved) : ")
  issues.append({"id":id, "title":title, "type":type, "priority":priority, "reporter":reporter, "status":status})
cnt=0
for i in range(5):
  if issues[i]["status"]=="Open":
    cnt+=1
    if cnt==1:
      issues[i]["priority"]="Critical"
print("-"*40)
print(f"Open issues     : {cnt}")
print("First issue → priority updated to Critical.")
print(f"Last two issues : {issues[3:]}")
print("-"*40)

reporters=set()
for i in range(5):
  reporters.add(issues[i]["reporter"])
print(f"reporters        : {reporters}")
tech_stack=set()
for i in range(4):
  tech_stack.add(contributors[i]["language"])
print(f"tech stack       : {tech_stack}")
tech_stack.add("C++")
tech_stack.discard("Rust")
print(f"union            : {reporters.union(tech_stack)}")
print(f"intersection     : {reporters.intersection(tech_stack)}     difference : {reporters.difference(tech_stack)}")
priorities=set()
for i in range(5):
  priorities.add(issues[i]["priority"])
if "Critical" in priorities:
  print("Critical present : YES — flag for immediate review.")
else:
  print("Critical present : NO — no immediate review needed.")
print("-"*40)

priority_count={}
for i in range(5):
  p=issues[i]["priority"]
  if p in priority_count:
    priority_count[p]+=1
  else:
    priority_count[p]=1
print(f"priority keys      : {priority_count.keys()}")
print(f"priority values    : {priority_count.values()}")

status_groups={}
for i in range(5):
  s=issues[i]["status"]
  if s in status_groups:
    status_groups[s].append(issues[i]["title"])
  else:
    status_groups[s]=[issues[i]["title"]]
print("Status Groups -> ",end="")
for s, t in status_groups.items():
  print(f"{s}: {t}", end=" | ")

reporters_dict={}
for i in range(5):
  r=issues[i]["reporter"]
  if r in reporters_dict:
    reporters_dict[r]+=1
  else:
    reporters_dict[r]=1
max=0
for key in reporters_dict:
  if reporters_dict[key]>max:
    max=reporters_dict[key]
    max_reporter=key
print()
print(f"Top reporter        : {max_reporter} ({max} issues)")
issues[0].pop('type')
print(f"After pop('type')   : {issues[0]}")
print("-"*40)

import os
folder_name="open_track"
if not os.path.exists(folder_name):
  os.mkdir(folder_name)
print(f"Folder created : {folder_name}/")
report_path = os.path.join(folder_name, "project_report.txt")
csv_path = os.path.join(folder_name, "issues.csv")
try:
  with open(report_path,"w") as f:
    f.write("="*40 + "\n")
    f.write(f"{project[0]} — PROJECT REPORT\n")
    f.write("="*40 + "\n")
    f.write(f"Project Info : {project}\n")
    f.write(f"Contributors : {contributors}\n")
    f.write(f"Issues : {issues}\n")
    f.write(f"Priority Count : {priority_count}\n")
    f.write(f"Top Reporter : {max_reporter}\n")
  with open(csv_path,"w") as f:
    f.write("id,title,priority,reporter,status\n")
    for i in issues:
      f.write(f"{i["id"]},{i["title"]},{i["priority"]},{i["reporter"]},{i["status"]}\n")
  print(f"Files saved : {os.listdir(folder_name)}")
except IOError:
  print("File write error")

try:
  with open(report_path,"r") as f:
    print("--- read() ---")
    print(f.read())
  with open(csv_path,"r") as f:
    print(f.read())
  with open(report_path, "r") as f:
    print("--- readline() ---")
    print(f"Line 1 : {f.readline().strip()}")
    print(f"Line 2 : {f.readline().strip()}")
  with open(csv_path, "r") as f:
    print("--- readlines() ---")
    lines = f.readlines()
    count = 0
    for line in lines:
      if "Critical" in line or "High" in line:
        print(line.strip())
        count += 1
    print(f"Total lines : {len(lines)}     Critical/High lines : {count}")
except FileNotFoundError:
  print("File not found")
print("="*40)
print(f"{project[0]} - FINAL SUMMARY ")
print("="*40)
print(f"Project : {project[0]}    Version : {project[1]}    Lead ; {project[4]}")
print(f"Contributors : {len(contributors)}  Names : {names}")
print(f"Tech Stack : {tech_stack}")
print(f"Issues : {len(issues)}  Open : {cnt}  Reporters : {len(reporters)}")
print(f"Top reporter : {max_reporter} ({max} issues)")
print(priority_count)

print(f"Report : {report_path}")
print(f"CSV    : {csv_path}")
print("="*40)
print(f"{project[0]} complete. Thank you for contributing to open source!")
print("="*40)

urgent = [i["title"] for i in issues if i["priority"] in ["Critical", "High"]]
print(f"Urgent issues : {urgent}  Length : {len(urgent)}")

with open(report_path, "a") as f:
    f.write("\n--- URGENT ISSUES ---\n")
    for title in urgent:
        f.write(f"{title}\n")

with open(report_path, "r") as f:
    lines = f.readlines()
    print("--- Last 6 lines ---")
    for line in lines[-6:]:
        print(line.strip())
```
